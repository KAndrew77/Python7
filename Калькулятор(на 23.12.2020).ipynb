```python
import math, cmath
import numpy as np
# используются методы array cbrt zeros real isreal
from functools import reduce

class VXQ:
    def __init__(self, inp):
        if "**" in inp:
            v, q = inp.split("**")[0].replace("x",""), inp.split("**")[1]
            if v and v != "+" and v != "-":
                self.v = float(v.replace("*",""))
            else:
                if v == "+": self.v = float(1)
                elif v == "-": self.v = float(-1)
                else: self.v = float(1)
            self.q = int(q)
        else:
            if "*x" in inp:
                self.v = float(inp.split("*")[0])
                self.q = int(1)
            elif "x" in inp:
                self.v = float(inp.replace("x","1"))
                self.q = int(1)
            else:
                self.v = float(inp)
                self.q = int(0)

    def __str__(self):
        if self.v == 0:
            return ""
        else:
            if self.v == 1:
                tempv = ""
            else:
                if self.q == 0:
                    tempv = str(self.v)
                else:
                    tempv = str(self.v)+"*"
            if self.q == 0:
                return f"{tempv}"
            elif self.q == 1:
                return f"{tempv}x"
            else:
                return f"{tempv}x**{self.q}"

def handler(inp):
    print("\n------------------------------")
    print("\nВведенное уравнение:")
    print(inp)
    inp = inp.replace("+ ", "+").replace("- ", "-")
    inpleft, inpright = inp.split(" = ")[0], inp.split(" = ")[1]
    inp = inpleft.split() + [("-"+i).replace("-+", "-").replace("--", "+") for i in inpright.split()]
    vars = [VXQ(i) for i in inp]
    varsq0 = [i for i in vars if i.q == 0]
    varsq1 = [i for i in vars if i.q == 1]
    varsq2 = [i for i in vars if i.q == 2]
    varsq3 = [i for i in vars if i.q == 3]
    varsq4 = [i for i in vars if i.q == 4]
    varq0 = VXQ(f"{sum([i.v for i in varsq0])}")
    varq1 = VXQ(f"{sum([i.v for i in varsq1])}*x")
    varq2 = VXQ(f"{sum([i.v for i in varsq2])}*x**2")
    varq3 = VXQ(f"{sum([i.v for i in varsq3])}*x**3")
    varq4 = VXQ(f"{sum([i.v for i in varsq4])}*x**4")
    vars = [varq4, varq3, varq2, varq1, varq0]
    vars = " + ".join([str(i) for i in vars if i.v])
    out = f"{vars} = 0"
    print("Упрощенное уравнение:")
    print(out)
    if varq4.v:
        print("Биквадратное уравнение")
        res = f_bs(varq4.v,varq2.v,varq0.v)
    elif varq3.v:
        print("Кубическое уравнение")
        res = f_c(varq3.v,varq2.v,varq1.v,varq0.v)
    elif varq2.v:
        print("Квадратное уравнение")
        res = f_s(varq2.v,varq1.v,varq0.v)
    elif varq1.v:
        print("Линейное уравнение")
        res = f_l(varq1.v,varq0.v)
    else:
        print("Неизвестный тип уравнения")
        res = "Пустое множество"
    return res

def f_l(a,b):
    print("\nax + b = 0")
    print(f"{a}x + {b} = 0")
    print("\nКоэффициенты уравнения:")
    print(f"a = {a}")
    print(f"b = {b}")

    x = -b / a
    print("\nОтвет:")
    print("x = %.2f" % x)
    return [x]

def f_s(a,b,c):
    print("\nax^2 + bx + c = 0")
    print(f"{a}x^2 + {b}x + {c} = 0")
    print("\nКоэффициенты уравнения:")
    print(f"a = {a}")
    print(f"b = {b}")
    print(f"c = {c}")

    if a == 0:
        print("В квадратном уравнении a не может равняться 0")
        return ["Пустое множество"]

    discr = b ** 2 - 4 * a * c
    print("Дискриминант D = %.2f" % discr)

    if discr > 0:
        x1 = (-b + math.sqrt(discr)) / (2 * a)
        x2 = (-b - math.sqrt(discr)) / (2 * a)
        print("\nОтвет:")
        print("x1 = %.2f \nx2 = %.2f" % (x1, x2))
        return [x1, x2]
    elif discr == 0:
        x = -b / (2 * a)
        print("\nОтвет:")
        print("x = %.2f" % x)
        return [x]
    else:
        print("\nОтвет:")
        print("Корней нет")
        return ["Пустое множество"]

def f_bs(a,b,c):
    print("\nax^4 + bx^2 + c = 0")
    print(f"{a}x^4 + {b}x^2 + {c} = 0")
    print("\nКоэффициенты уравнения:")
    print(f"a = {a}")
    print(f"b = {b}")
    print(f"c = {c}")

    if a == 0:
        print("В биквадратном уравнении a не может равняться 0")
        return ["Пустое множество"]

    discr = b ** 2 - 4 * a * c
    print("Дискриминант D = %.2f" % discr)

    if discr > 0:
        x1_2 = (-b + math.sqrt(discr)) / (2 * a)
        x2_2 = (-b - math.sqrt(discr)) / (2 * a)
        print("x1^2 = %.2f \nx2^2 = %.2f" % (x1_2, x2_2))
        x1 = math.sqrt(x1_2)
        x2 = math.sqrt(x1_2)*-1
        x3 = math.sqrt(x2_2)
        x4 = math.sqrt(x2_2)*-1
        print("\nОтвет:")
        print("x1 = %.2f \nx2 = %.2f \nx3 = %.2f \nx4 = %.2f" % (x1, x2, x3, x4))
        return [x1, x2, x3, x4]
    elif discr == 0:
        x_2 = -b / (2 * a)
        print("x^2 = %.2f" % x_2)
        x1 = math.sqrt(x_2)
        x2 = math.sqrt(x_2)*-1
        print("\nОтвет:")
        print("x1 = %.2f \nx2 = %.2f" % (x1, x2))
        return [x1, x2]
    else:
        print("\nОтвет:")
        print("Корней нет")
        return ["Пустое множество"]

def f_c(a,b,c,d):
    print("\nax^3 + bx^2 + cx + d = 0")
    print(f"{a}x^3 + {b}x^2 + {c}x + {d} = 0")
    print("\nКоэффициенты уравнения:")
    print(f"a = {a}")
    print(f"b = {b}")
    print(f"c = {c}")
    print(f"d = {d}")

    if a == 0:
        print("В кубическом уравнении a не может равняться 0")
        return ["Пустое множество"]

    a = np.array([a])
    b = np.array([b])
    c = np.array([c])
    d = np.array([d])
    a,b,c,d = a+0j, b+0j, c+0j, d+0j
    all_ = (a != np.pi)

    Q = (3*a*c - b**2)/ (9*a**2)
    R = (9*a*b*c - 27*a**2*d - 2*b**3) / (54 * a**3)
    D = Q**3 + R**2
    S = 0
    if np.isreal(R + cmath.sqrt(D)):
        S = np.cbrt(np.real(R + cmath.sqrt(D)))
    else:
        S = (R + cmath.sqrt(D))**(1/3)
    T = 0
    if np.isreal(R - cmath.sqrt(D)):
        T = np.cbrt(np.real(R - cmath.sqrt(D)))
    else:
        T = (R - cmath.sqrt(D))**(1/3)
    result = np.zeros(tuple(list(a.shape) + [3])) + 0j
    result[all_,0] = - b / (3*a) + (S+T)
    result[all_,1] = - b / (3*a)  - (S+T) / 2 + 0.5j * cmath.sqrt(3) * (S - T)
    result[all_,2] = - b / (3*a)  - (S+T) / 2 -  0.5j * cmath.sqrt(3) * (S - T)

    if round(result[0][0].imag, 14) != 0:
        x1=round(result[0][0].real, 14) + round(result[0][0].imag, 14) * 1j
    else:
        x1=round(result[0][0].real, 14)
    if round(result[0][1].imag, 14) != 0:
        x2=round(result[0][1].real, 14) + round(result[0][1].imag, 14) * 1j
    else:
        x2=round(result[0][1].real, 14)
    if round(result[0][2].imag, 14) != 0:
        x3=round(result[0][2].real, 14) + round(result[0][2].imag, 14) * 1j
    else:
        x3=round(result[0][2].real, 14)
    print("\nОтвет:")
    print(f"x1 = {x1} \nx2 = {x2} \nx3 = {x3}")
    return [x1, x2, x3]

def main():
    print("1 - Ввести систему уравнений\n2 - Считать систему уравнений из файла inp.txt")
    cmd = int(input("Ввод: "))
    if cmd == 1:
        print("\nПример ввода системы уравнений\nВвод: x**2 = + x + 6\nВвод: 2*x**3 + 8*x**2 - 8*x - 32 = 0\nВвод: .\n\nЧтобы завершить систему введите точку")
        inps = []
        inp = input("Ввод: ")
        while inp != ".":
            inps.append(inp)
            inp =  input("Ввод: ")
    elif cmd == 2:
        with open("inp.txt", "r", encoding="utf-8") as f:
            inps = f.readlines()
    ress = []
    for inp in inps:
        if inp != ".":
            res = handler(inp)
            ress.append(res)
    print("\n------------------------------")
    print("\nВсе ответы:")
    print(ress)
    res = list(reduce(lambda x,y: set(x) & set(y), ress))
    print("\n------------------------------")
    print("\nИтоговый ответ:")
    print(res)

if __name__ == '__main__':
    main()
```
